#Analysing number of agents with particular Agent tone

```python
import matplotlib.pyplot as plt
# Count the frequency of each tone
tone_counts = sentiment_data['agent_tone'].value_counts()

# Create a bar plot of the tone frequencies
plt.figure(figsize=(10, 6))
sns.countplot(x='agent_tone', data=sentiment_data)
plt.title('Frequency of Agent Tones')
plt.xlabel('Agent Tone')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

# Display the counts on top of each bar
for i, count in enumerate(tone_counts):
  plt.text(i, count, str(count), ha='center', va='bottom')

plt.show()
```

##Analysing number of customers with particular customer tone

```python
unique_tones = sentiment_data['customer_tone'].unique()
print("Unique Customer Tones:")
for tone in unique_tones:
  print(tone)
# Count the frequency of each tone
tone_counts = sentiment_data['customer_tone'].value_counts()

# Create a bar plot of the tone frequencies
plt.figure(figsize=(10, 6))
sns.barplot(x=tone_counts.index, y=tone_counts.values)
plt.xlabel('Customer Tone')
plt.ylabel('Frequency')
plt.title('Frequency of Customer Tones')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

# Add exact counts on top of each bar
for p in plt.gca().patches:
    plt.gca().text(p.get_x() + p.get_width() / 2., p.get_height(),
                 '{:.0f}'.format(p.get_height()),
                 ha='center', va='bottom')

plt.show()
```

#Anaylsing range of AHT and AST

```python


print(f"Maximum AHT: {data['AHT'].max()}")
print(f"Minimum AHT: {data['AHT'].min()}")
print(f"Maximum AST: {data['AST'].max()}")
print(f"Minimum AST: {data['AST'].min()}")
```

```python
# Convert datetime columns from object type to datetime type
data['call_start_datetime'] = pd.to_datetime(data['call_start_datetime'])
data['agent_assigned_datetime'] = pd.to_datetime(data['agent_assigned_datetime'])
data['call_end_datetime'] = pd.to_datetime(data['call_end_datetime'])

# Create the AHT (Average Handle Time) column:
# Time from agent assignment to the end of the call
data['AHT'] = (data['call_end_datetime'] - data['agent_assigned_datetime']).dt.total_seconds()

# Create the AST (Average Speed to Answer) column:
# Time from call start to agent assignment
data['AST'] = (data['agent_assigned_datetime'] - data['call_start_datetime']).dt.total_seconds()

# Check the new columns
print(data[['AHT', 'AST']].describe())

# Optional: Save the dataset with the new columns
data.to_csv('call_center_data_with_AHT_AST.csv', index=False)
```

#Correlation between Agent tone and customer tone

```python


import pandas as pd
import matplotlib.pyplot as plt

tone_crosstab = pd.crosstab(sentiment_data['agent_tone'], sentiment_data['customer_tone'])

# Plot the correlation using a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(tone_crosstab, annot=True, fmt="d", cmap="viridis", cbar=True)
plt.title("Correlation between Agent Tone and Customer Tone")
plt.xlabel("Customer Tone")
plt.ylabel("Agent Tone")
plt.show()
```

```python
# Merging all dataframes
merged_df = data.merge(sentiment_data, on=['call_id', 'agent_id'], how='left')\
                   .merge(reason_data, on='call_id', how='left')\
                   .merge(customer_data, on='customer_id', how='left')

# Preview merged data
print(merged_df.head())
```

```python


# Save the merged DataFrame to a CSV file in your Google Drive
merged_df.to_csv('/content/drive/MyDrive/SkyHack Dataset/merged_data.csv', index=False)
```

#AHT AND AST VS customer_elite_level_code

```python


import matplotlib.pyplot as plt
# Group by customer loyalty status and calculate the mean AHT and AST
loyalty_aht_ast = merged_df.groupby('elite_level_code').agg({'AHT': 'mean', 'AST': 'mean'})

# Display the results
print(loyalty_aht_ast)

# Create a bar plot to visualize the mean AHT and AST by customer loyalty status
loyalty_aht_ast.plot(kind='bar', figsize=(10, 6))
plt.title('Mean AHT and AST by Customer Loyalty Status')
plt.xlabel('Customer Loyalty Status')
plt.ylabel('Mean Time (seconds)')
plt.xticks(rotation=0)
plt.legend(['Mean AHT', 'Mean AST'])
plt.show()
```
